```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scipy.stats as sps
%matplotlib inline
```

```python
test_sizes = list(range(1, 131))
data = np.loadtxt("iris-results.txt")
data = data * 100.0 / 10
means = np.mean(data,0)
stderr = np.std(data,0) / np.sqrt(data.shape[0])
plt.bar([str(x) for x in test_sizes],means,yerr=stderr,capsize=2)
plt.xticks(np.arange(min(test_sizes), max(test_sizes)+1, 10.0))
plt.xlabel("Number of Centroids [N]")
plt.ylabel("Mean Accuracy [%]")
plt.title("Iris Dataset")
plt.show()
```

```python
test_sizes = list(range(1, 96))
data = np.loadtxt("cancer-results.txt")
data = data * 100.0 / 10
means = np.mean(data,0)
stderr = np.std(data,0) / np.sqrt(data.shape[0])
plt.bar([str(x) for x in test_sizes],means,yerr=stderr,capsize=2)
plt.xticks(np.arange(min(test_sizes), max(test_sizes)+1, 10.0))
plt.xlabel("Number of Centroids [N]")
plt.ylabel("Mean Accuracy [%]")
plt.title("Cancer Dataset")
plt.show()
```

```python
data = np.loadtxt("iris-results.txt")
means = np.apply_along_axis(np.mean,0,data)*100.0/10.0
stderr = np.apply_along_axis(np.std,0,data)*100.0/(10.0*np.sqrt(np.shape(data)[0]-1))
stderr = np.apply_along_axis(sps.sem,0,data)*100.0/10.0
plt.figure(figsize=(15,5))
plt.errorbar(np.array(range(len(means)))+1.0,means,yerr=stderr)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Percentage Correct (%) [+- Std. Err.]')
```
